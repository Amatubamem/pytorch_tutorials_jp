```python
import numpy as np
import matplotlib.pyplot as plt

# xの値の範囲を定義
x = np.linspace(0, 1, 100)  # -2πから2πまでの範囲を100個の点でサンプリング

rho = 0.3

# 各関数のy値を計算
y1 = rho * np.log(rho/x) + (1-rho)*np.log((1-rho)/(1-x))
rho = 0.5
y2 = rho * np.log(rho/x) + (1-rho)*np.log((1-rho)/(1-x))
y3 = (x-rho)**2

# グラフを描画
plt.figure(figsize=(8, 6))  # グラフのサイズを調整

plt.plot(x, y1, label='KL(rho=0.3)', linestyle='-', color='blue')
plt.plot(x, y2, label='KL(rho=0.5)', linestyle='--', color='red')
plt.plot(x, y3, label='drho^2(rho=0.5)', linestyle='-.', color='green')

plt.xlabel('rho_j')
plt.ylabel('diff')
plt.title('KL')
plt.legend()  # 凡例を表示

plt.grid(True)  # グリッドを表示
plt.show()  # グラフを表示
```
